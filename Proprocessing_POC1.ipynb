```python
import re

# Step 1: Load transcript from your uploaded notepad file
file_path = "team_meeting_transcript_hindi (1).txt"   # <-- change to your actual filename
with open(file_path, "r", encoding="utf-8") as f:
    transcript = f.read()

# Step 2: Preprocessing function
def preprocess_transcript(text):
    # 1. Remove timestamps [hh:mm:ss]
    text = re.sub(r'\[\d{2}:\d{2}:\d{2}\]', '', text)
    
    # 2. Remove speaker names (anything like "Name:")
    text = re.sub(r'\s*[A-Za-z]+\s*:', '', text)
    
    # 3. Remove extra newlines and spaces
    text = re.sub(r'\n+', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    
    # 4. Strip leading/trailing spaces
    text = text.strip()
    
    return text

# Step 3: Apply preprocessing
clean_transcript = preprocess_transcript(transcript)

# Step 4: Save preprocessed transcript (optional)
with open("clean_transcript.txt", "w", encoding="utf-8") as f:
    f.write(clean_transcript)

print("---- Preprocessed Transcript ----")
print(clean_transcript[:])  # show first 500 characters
```
